Karena kita akan menampilkan kualitas udara maka dibuatkan sebuah fungsi yang akan mengelompokkan kualitas udara berdasarkan faktor komponen PM2.5, PM10, SO2, NO2, CO, dan O3

```python
#pada fungsi ini akan membuat batasan sebanyak 3 jangkauan
def air_quality_range(PM25, PM10, SO2, NO2, CO, O3):
    standar ={
        'PM25' : [(0, 15), (15, 65), (65, float('inf'))],
        'PM10' : [(0, 50), (50, 150), (150, float('inf'))],
        'SO2' : [(0, 20), (20, 150), (150, float('inf'))],
        'NO2' : [(0, 40), (40, 150), (150, float('inf'))],
        'CO' : [(0, 1000), (1000, 5000), (5000, float('inf'))],
        'O3' : [(0, 100), (100, 180), (180, float('inf'))]
    }
    #fungsi menentukan kategori kualitas udara
    def air_quality_label(value, param):
        for quality, bound in enumerate(standar[param], start=1):
            if isinstance(bound, tuple):
                if bound[0] <= value < bound[1]:
                    return quality
            else:
                if value == bound:
                        return quality
        return 3
    '''
    Fungsi air_quality_label bertujuan untuk menentukan kategori kualitas udara
    untuk setiap parameter berdasarkan standar baku mutu udara ambien 
    yang disimpan dalam variabel standar. Fungsi ini mengembalikan nilai numerik antara 1 sampai 
    3 yang mewakili kategori kualitas udara yaitu Baik, Sedang, Tidak Sehat
    '''
    PM25_quality = air_quality_label(float(PM25), 'PM25')
    PM10_quality = air_quality_label(float(PM10), 'PM10')
    SO2_quality = air_quality_label(float(SO2), 'SO2')
    NO2_quality = air_quality_label(float(NO2), 'NO2')
    CO_quality = air_quality_label(float(CO), 'CO')
    O3_quality = air_quality_label(float(O3), 'O3')

    # Menghitung rata-rata kualitas 
    quality_avg = (PM25_quality + PM10_quality + SO2_quality + NO2_quality + CO_quality + O3_quality)/6

    if quality_avg < 2:
        return 'Good'
    elif quality_avg < 2.5:
        return 'Moderate'
    else:
        return 'Unhealthy'

#menerapkan fungsi air_quality_range pada setiap baris data frame df_aq
#dan menyimpan hasilnya dalam kolom baru bernama air_quality.
df_aq['air_quality'] = df_aq.apply(
    lambda row: air_quality_range(
        row['PM2.5'],
        row['PM10'],
        row['SO2'],
        row['NO2'],
        row['CO'],
        row['O3']), axis=1)
#Mengubah tipe data kolom air_quality menjadi kategori.
df_aq['air_quality'] = df_aq.air_quality.astype('category')
            
```

Pertanyaan 2: Wilayah yang memiliki kualitas udara terbaik dan terburuk periode 2013 - 2017?

```python
#menyiapkan dataframe yang berisi aqi berdasarkan kolom station
station_df = df_aq.groupby(by="station").aqi.sum().sort_values(ascending=False).reset_index()

# membuat 2 kanvas untuk visualisasi data best & worst air quality by station
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

#memberi warna bar
colors1 = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
colors2 = ["#E32925", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="aqi", y="station", data=station_df.head(5), palette=colors1, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Air Quality Index (AQI)", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
 
sns.barplot(x="aqi", y="station", data=station_df.sort_values(by="aqi", ascending=True).head(5), palette=colors2, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Air Quality Index (AQI)", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Best and Worst Air Quality Index (AQI)\nby Station", fontsize=20)
plt.show()
```

Visualisasi data diatas menunjukkan Indeks Kualitas Udara terbaik berada pada station Dingling , sedangkan Indeks Kualitas Udara terburuk berada pada station Nongzhanguan.
